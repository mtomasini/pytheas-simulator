# Playing with the A*

```python
import heapq
from typing import Dict, Tuple, List, Iterator, Optional, TypeVar

Var = TypeVar('Var')
Position = Tuple[float, float]

class PriorityQueue:
    """Class to define PriorityQueue (used in Astar).

    Attributes:
        elements (List[Tuple[float, T]]): elements a queue used to check points in A-star algorithm.
    Methods:
        empty(): return True if self.elements is empty.
        put(item, priority): insert new element in priority list
        get: return weight for iteration step
    """
    def __init__(self):
        self.elements: List[Tuple[float, Var]] = []
    
    def empty(self) -> bool:
        return not self.elements
    
    def put(self, item: Var, priority: float):
        heapq.heappush(self.elements, (priority, item))
    
    def get(self) -> Var:
        return heapq.heappop(self.elements)[1]
```

```python
def find_nearest_index(map, launching_site):
    id_lat = min(enumerate(map.latitude.values), key=lambda x: abs(x[1]-launching_site[0]))
    id_lon = min(enumerate(map.longitude.values), key=lambda x: abs(x[1]-launching_site[1]))
    
    return id_lat[0], id_lon[0]
    # min(np.abs(launching_site[0] - skagerrak_map.currents_data.uo.latitude.values))
```
